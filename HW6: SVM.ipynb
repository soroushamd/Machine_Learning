```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import scipy.io as sio
from scipy import optimize
from sklearn import svm
```

```python
data = sio.loadmat('data/ex6data1.mat')
# print data
X = data['X']
X1 = X[:,0]
X2 = X[:,1]
Y = data['y']

print X.shape, Y.shape

fig, ax = plt.subplots()
# ax.plot(X, Y, 'x', markeredgewidth=2)
plt.scatter(X1, X2, c=Y[:,0])#, cmap=plt.cm.autumn)
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
```

```python
clf_lin1 = svm.SVC(kernel='linear', C=1)
clf_lin10 = svm.SVC(kernel='linear', C=10)
```

```python
def plotDecLine(svm,X,Y):
    
    svm.fit(X, Y)  
    
    X1 = X[:,0]
    X2 = X[:,1]
    x1 = np.linspace(np.min(X1),np.max(X1),100)
    x2 = np.linspace(np.min(X2),np.max(X2),100)
    xx, yy = np.meshgrid(x1,x2)
#     print xx.shape
    xx_ = xx.flatten()
    yy_ = yy.flatten()

    xxyy = np.array([xx_, yy_]).T
#     print xxyy.shape
    predicted = svm.predict(xxyy)

    predicted = predicted.reshape(100,100)

    CS = plt.contour(x1,x2,predicted, levels=0.5)
    plt.scatter(X1, X2, c=Y[:,0])#, cmap=plt.cm.autumn)   
    
    plt.show()
    
    return 
```

```python
plotDecLine(clf_lin1,X,Y)
plotDecLine(clf_lin10,X,Y)
```

```python
data = sio.loadmat('data/ex6data3.mat')
# print data
X = data['X']
Xval = data['Xval']
X1 = X[:,0]
X2 = X[:,1]
Y = data['y']
Yval = data['yval']

print X.shape, Y.shape

fig, ax = plt.subplots()
# ax.plot(X, Y, 'x', markeredgewidth=2)
plt.scatter(X1, X2, c=Y[:,0])#, cmap=plt.cm.autumn)
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
```

```python
gamma = np.linspace(0.1,50,10)
c = np.linspace(0.1,50,10)

scores = np.zeros((10,10))

for i in xrange(10):
    for j in xrange(10):
        clf = svm.SVC(kernel='rbf', gamma=gamma[i], C=c[j])
        clf.fit(X, Y) 
        scores[i,j] = clf.score(Xval,Yval)
```

```python
print "max score is ", scores[np.where(scores==scores.max())][0]
```

```python
a = np.where(scores==scores.max())
print a[0][0], a[1][0], scores[a[0][0], a[1][0]]
```

```python
clf = svm.SVC(kernel='rbf', gamma=gamma[a[0][0]], C=c[a[1][0]])
plotDecLine(clf,X,Y)
```
