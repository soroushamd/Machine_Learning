```python
def parameterEstimateUni(X):
    return np.mean(X,axis=0),np.var(X,axis=0)
def parameterEstimateMulti(X):
    return np.mean(X,axis=0),np.cov(X.T)
```

```python
def calcEps(X,Xval,eps):
    
#     means_m,cov = parameterEstimateMulti(X)
#     p = multivariate_normal.pdf(X, means_m, cov)

    means,sigmas = parameterEstimateUni(X)
    p = np.ones(Xval.shape[0])
    for i in xrange(Xval.shape[1]):
        p = p*multivariate_normal.pdf(Xval[:,i], means[i], sigmas[i])
#     p*=np.sqrt(2.*3.141592)
    
    tmp = np.fmin(p, eps)
    tp = float(np.sum(Y)) # total true pos
    fp = float(np.sum(1-Y[np.where(tmp!=eps)])) # total false positive detected
    fn = np.maximum(tp-np.sum(Y[np.where(tmp!=eps)]),0) # total true positives we missed
#     print p[np.where(tmp!=eps)]
#     print tp,fp,fn
    pres = tp/(tp+fp)
    rec = tp/(tp+fn)
#     print pres, rec
    F1 = 2.*pres*rec/(pres+rec)
        
    return F1, (tmp != eps).sum() # score and # of anomalies
```

```python
eps = np.linspace(0.01,1e-5,100)
F1 = np.zeros(eps.shape)
for i in xrange(100):
    F1[i] = calcEps(X,Xval,eps[i])[0]
    
print np.max(F1)
# calcEps(X,Xval,9e-5)
```

```python
def cost(Xtheta,Y,R,lambda_,nf=100,nm=R.shape[0],nu=R.shape[1]):
    
    """
    Function to calculate the cost function. Returns a scalar.
    Xtheta is a combined flattened arrays of X and theta, we reshape it before using it here
    Arguments:
    """ 
    
    X = Xtheta[:nm*nf].reshape(nm,nf)
    theta = Xtheta[nm*nf:].reshape(nu,nf)
    
    tmp = np.dot(X,theta.T)*R-Y
    return 0.5 * np.sum((tmp[tmp!=0])**2) + 0.5*lambda_ * (np.sum(theta**2.) + np.sum(X**2.))
```

```python
np.random.seed(1)
X = np.random.uniform(0,1,(nm,nf))
theta = np.random.uniform(0,1,(nu,nf))
Xtheta = np.concatenate((X.flatten(),theta.flatten()))
cost(Xtheta,Y,R,10)
```

```python
def costGrad(Xtheta,Y,R,lambda_=0.,nf=100,nm=R.shape[0],nu=R.shape[1]):
    
    """
    Function to calculate gradient of cost function w.r.t X and theta. Returns gradient of Xtheta.
    Xtheta is a combined flattened arrays of X and theta, we reshape it before using it here.
    Arguments:
    """   
    
    X = Xtheta[:nm*nf].reshape(nm,nf)
    theta = Xtheta[nm*nf:].reshape(nu,nf)
    
    tmp = np.dot(X,theta.T)*R-Y
    gradX = np.dot(tmp,theta) + lambda_*X
    gradT = np.dot(tmp.T,X) + lambda_*theta
    
    return np.concatenate((gradX.flatten(),gradT.flatten()))
```

```python
def gradCheck(n,Y,R,lambda_=0.,nf=100,nm=R.shape[0],nu=R.shape[1]):
    
    """
    Function to verify proper gardient calculation
    It randomly initializes X and theta matrices, calculate gradient w.r.t X and theta @ entry "n" in Xtheta array.
    Arguments:
    """ 

    eps = 1e-4
    X = np.random.uniform(0,1,(nm,nf))
    theta = np.random.uniform(0,1,(nu,nf))
    Xtheta = np.concatenate((X.flatten(),theta.flatten()))
    grad = costGrad(Xtheta,Y,R,lambda_,nm=R.shape[0],nu=R.shape[1],nf=100)
    Xtheta[n] = Xtheta[n]-eps
    tmp1 = cost(Xtheta,Y,R,lambda_,nm=R.shape[0],nu=R.shape[1],nf=100)
    Xtheta[n] = Xtheta[n]+2.*eps
    tmp2 = cost(Xtheta,Y,R,lambda_,nm=R.shape[0],nu=R.shape[1],nf=100)
    
    return (tmp2-tmp1)/(2.*eps), grad[n]
```

```python
print "these two numbers must be almost equal"
gradCheck(100,Y,R,10.,nm=R.shape[0],nu=R.shape[1],nf=100)
```

```python
np.random.seed(1)
X = np.random.uniform(0,1,(nm,nf))
theta = np.random.uniform(0,1,(nu,nf))
x0 = np.concatenate((X.flatten(),theta.flatten()))

result = scipy.optimize.fmin_cg(cost, x0=x0, fprime=costGrad, \
                               args=(Y,R,5.), \
                                maxiter=50,disp=True,full_output=True)
```

```python
### lets make predictions using our trained model and sort my top recommendations
X = result[0][:nm*nf].reshape(nm,nf)
theta = result[0][nm*nf:].reshape(nu,nf)
recommendation = np.dot(X,theta.T)
sorted_recommendation = np.argsort(recommendation[:,0])
# sorted_recommendation
```
